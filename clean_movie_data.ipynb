```python
import pandas as pd
```

```python
df = pd.read_csv("titles.csv")
df.head(20)
```

```python
mask = df['type'] == 'SHOW'
df = df[~mask]
```

```python
df.head(20)
```

```python
regex = r'[^\x00-\x7F]+'
mask = df['title'].str.contains(regex)

df = df[~mask]
```

```python
df.to_csv("titles_cleaned.csv", index=False)
```
